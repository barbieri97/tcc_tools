```python
import pandas as pd
```

```python
pubm = pd.read_csv('pubmed.csv')
new_pubm = pubm[['Title', 'Authors', 'DOI']]
new_pubm.insert(3, 'base', 'pubmed')
new_pubm.columns = ['Title', 'Authors', 'DOI', 'Base']
new_pubm
```

```python
bvs = pd.read_csv('bvs.csv')
new_bvs = bvs[['Title', 'Authors', 'DOI']]
new_bvs.insert(3, 'base', 'bvs')
new_bvs.columns = ['Title', 'Authors', 'DOI', 'Base']
new_bvs
```

```python
psycnet = pd.read_csv('PsycNET.csv')
new_psycnet = psycnet[['DocumentTitle', 'Author', 'DOI']]
new_psycnet.insert(3, 'base', 'psycnet')
new_psycnet.columns = ['Title', 'Authors', 'DOI', 'Base']
new_psycnet
```

```python
rbtcc = pd.read_csv('rbtcc.csv')
rbtcc.insert(3, 'base', 'rbtcc')
rbtcc.columns = ['Title', 'Authors', 'DOI', 'Base']
rbtcc
```

```python
acta = pd.read_csv('acta_comportamentalia.csv')
acta.insert(3, 'base', 'acta')
acta.columns = ['Title', 'Authors', 'DOI', 'Base']
acta
```

```python
all_articles = pd.concat([acta, rbtcc, new_bvs, new_psycnet, new_pubm], ignore_index=True, names=['Title', 'Author', 'DOI', 'Base'])
all_articles.to_csv('/home/barbieri/projetos/tools_tcc/all_articles.csv')

teste = pd.read_csv('all_articles.csv')
teste[teste['Title']=='Clinical and cost-effectiveness of an intervention for reducing cholesterol and cardiovascular risk for people with severe mental illness in English primary care: a cluster randomised controlled trial'].index.tolist()
```
